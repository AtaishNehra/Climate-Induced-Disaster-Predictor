```python
import pandas as pd
```

```python
sea_levels = pd.read_csv('Change_in_Mean_Sea_Levels.csv')
sea_levels.head()
```

```python
#print(sea_levels.dtypes)
dates = sea_levels['Date'].str.split('/', expand = True)
sea_levels['Year'] = dates[2]
sea_levels.head()
```

```python
sea_levels = sea_levels.drop(columns = ['ObjectId', 'ISO2', 'ISO3', 'Indicator', 'Unit', 'Source', 'CTS_Code', 'CTS_Name', 'CTS_Full_Descriptor', 'Date']).reset_index()
sea_levels.head()
```

```python
#creating only the world average
avg_sea_levels = sea_levels.drop(columns = ['Measure','index'])
avg_sea_level = avg_sea_levels.groupby(['Year', 'Country']).mean().reset_index()
avg_sea_level.head()
```
