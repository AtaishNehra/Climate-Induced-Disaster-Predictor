# CO Concentrations

```python
import pandas as pd

# Load the dataset
file_path = "C:/Users/ATAISH NEHRA/Downloads/Cleaned Datasets/co_concentrations.csv"
co_concentrations_df = pd.read_csv(file_path)

# Display the first few rows of the dataset
co_concentrations_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting up the environment for visualization
sns.set(style="whitegrid")

# Trend Analysis: Plotting the trends in CO PPM and CO Percent over time
fig, ax1 = plt.subplots(figsize=(12, 6))

color = 'tab:red'
ax1.set_xlabel('Time (Years)')
ax1.set_ylabel('CO PPM', color=color)
ax1.plot(co_concentrations_df.index, co_concentrations_df['CO PPM'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:blue'
ax2.set_ylabel('CO Percent', color=color)  # we already handled the x-label with ax1
ax2.plot(co_concentrations_df.index, co_concentrations_df['CO Percent'], color=color)
ax2.tick_params(axis='y', labelcolor=color)
```

Trend Analysis: The first plot shows the trends in CO PPM (in red) and CO Percent (in blue) over time. Each line represents the change in these values across the dataset's timespan. This visualization is crucial for understanding the long-term trends in carbon monoxide levels, which can be a significant factor in predicting climate-induced disasters.

```python
# Distribution Analysis: Histograms for CO PPM and CO Percent
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
sns.histplot(co_concentrations_df['CO PPM'], kde=True, ax=axs[0])
axs[0].set_title('Distribution of CO PPM')
sns.histplot(co_concentrations_df['CO Percent'], kde=True, ax=axs[1])
axs[1].set_title('Distribution of CO Percent')
```


Distribution Analysis: The second set of plots are histograms showing the distribution of CO PPM (left) and CO Percent (right). These histograms provide insights into the spread and central tendencies of the carbon monoxide measurements. Understanding the distribution is important for neural network models, as it can impact how the model processes and learns from the data.

# Surface Temperature

```python
# Load the new dataset
surface_temp_file_path = "C:/Users/ATAISH NEHRA/Downloads/Cleaned Datasets/surface_temp_clean.csv"
surface_temp_df = pd.read_csv(surface_temp_file_path)

# Display the first few rows of the dataset to understand its structure
surface_temp_df.head()
```

```python
# Preparing data for global trend analysis
# Calculating the global average temperature anomaly for each year
surface_temp_df['Global Average'] = surface_temp_df.iloc[:, 1:].mean(axis=1)

# Global Trend Analysis: Plotting the global average temperature anomaly over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Global Average', data=surface_temp_df)
plt.title('Global Average Temperature Anomaly Over Time')
plt.xlabel('Year')
plt.ylabel('Temperature Anomaly (°C)')
plt.grid(True)
plt.show()
```

```python
# Selected countries list with the appropriate names from the dataset
selected_countries = ['United States', 'Brazil', 'India', 'Australia', 'China, P.R.: Mainland']

# Preparing the dataset for the country-specific analysis
country_specific_df = surface_temp_df[['Year'] + selected_countries]

# Plotting the temperature anomalies of the selected countries over time
plt.figure(figsize=(12, 6))
for country in selected_countries:
    sns.lineplot(x='Year', y=country, data=country_specific_df, label=country)
plt.title('Temperature Anomalies for Selected Countries Over Time')
plt.xlabel('Year')
plt.ylabel('Temperature Anomaly (°C)')
plt.legend()
plt.grid(True)
plt.show()
```

Global Trend Analysis: A line plot showing the global average temperature anomaly over time, illustrating the overall trend in global warming.

Country-Specific Analysis: A line plot comparing the temperature anomalies over time for the United States, Brazil, India, Australia, and Mainland China, highlighting regional differences in temperature changeear.

# Sea level changes

```python
# Load the new dataset
sea_level_file_path = "C:/Users/ATAISH NEHRA/Downloads/Cleaned Datasets/sea_level_changes.csv"
sea_level_df = pd.read_csv(sea_level_file_path)

# Display the first few rows of the dataset to understand its structure
sea_level_df.head()
```

```python
# Preparing data for trend analysis and global sea level change analysis
# Assuming the rows are ordered chronologically, we will add a 'Year' column for visualization purposes
sea_level_df['Year'] = range(1, len(sea_level_df) + 1)

# Selecting a few seas and oceans for trend analysis
selected_seas = ['Atlantic Ocean', 'Indian Ocean', 'Pacific Ocean', 'Southern Ocean', 'Caribbean Sea']

# Trend Analysis: Plotting the sea level changes of selected seas and oceans over time
plt.figure(figsize=(12, 6))
for sea in selected_seas:
    sns.lineplot(x='Year', y=sea, data=sea_level_df, label=sea)
plt.title('Sea Level Changes for Selected Seas and Oceans Over Time')
plt.xlabel('Year')
plt.ylabel('Sea Level Change (mm)')
plt.legend()
plt.grid(True)
plt.show()
```

Trend Analysis: The first line plot shows the sea level changes over time for selected seas and oceans, including the Atlantic Ocean, Indian Ocean, Pacific Ocean, Southern Ocean, and Caribbean Sea. This visualization helps in understanding the trends and fluctuations in sea levels in these regions.sasters.

```python
# Global Sea Level Change Analysis: Calculating and plotting the global average sea level change over time
sea_level_df['Global Average'] = sea_level_df.iloc[:, :-2].mean(axis=1)
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Global Average', data=sea_level_df)
plt.title('Global Average Sea Level Change Over Time')
plt.xlabel('Year')
plt.ylabel('Global Average Sea Level Change (mm)')
plt.grid(True)
plt.show()
```

Global Sea Level Change Analysis: The second line plot illustrates the global average sea level change over time, calculated across all seas and oceans in the dataset. This plot provides a broader view of how sea levels have been changing globally.

```python
# Heatmap Analysis: Creating a heatmap for the dataset
heatmap_data = sea_level_df.drop('Year', axis=1).transpose()
heatmap_data.columns = sea_level_df['Year']
plt.figure(figsize=(10, 15))
sns.heatmap(heatmap_data, cmap='coolwarm', yticklabels=True)
plt.title('Heatmap of Sea Level Changes')
plt.ylabel('Sea/Ocean')
plt.xlabel('Year')
plt.show()
```

Heatmap Analysis: The heatmap visualizes sea level changes across different seas and oceans over the years. This comprehensive view allows for easy comparison of sea level changes across regions and over time.

```python
# Distribution Analysis: Histograms for selected seas and oceans
fig, axs = plt.subplots(1, len(selected_seas), figsize=(20, 4), sharey=True)
for i, sea in enumerate(selected_seas):
    sns.histplot(sea_level_df[sea], kde=True, ax=axs[i])
    axs[i].set_title(f'Distribution of {sea}')
plt.suptitle('Distribution of Sea Level Changes for Selected Seas and Oceans')
plt.show()
```

Distribution Analysis: The histograms show the distribution of sea level changes for the selected seas and oceans. These plots are useful for understanding the spread and typical range of sea level changes in these regions.

These visualizations offer valuable insights into sea level changes, which are crucial for understanding and predicting the impact of climate change on different regions around the world. They also provide essential data that can be used in building neural network models for predicting climate-induced changes and disasters.

# Climate Disasters

```python
# Load the new dataset
climate_disasters_file_path = "C:/Users/ATAISH NEHRA/Downloads/Cleaned Datasets/climate_disasters_clean.csv"
climate_disasters_df = pd.read_csv(climate_disasters_file_path)

# Display the first few rows of the dataset to understand its structure
climate_disasters_df.head()
```

```python
# Trend Analysis: Plotting the trends in different types of climate disasters over time
disaster_types = ['Drought', 'Extreme temperature', 'Flood', 'Landslide', 'Storm', 'Wildfire']

plt.figure(figsize=(12, 6))
for disaster in disaster_types:
    sns.lineplot(x='Year', y=disaster, data=climate_disasters_df, label=disaster)
plt.title('Trends in Different Types of Climate Disasters Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Disasters')
plt.legend()
plt.grid(True)
plt.show()
```

Trend Analysis: The first line plot shows the trends in different types of climate disasters (Drought, Extreme Temperature, Flood, Landslide, Storm, Wildfire) over time. This visualization helps in understanding how the frequency of each type of disaster has changed over the years. type.

```python
# Total Disaster Trend Analysis: Analyzing the trend in the total number of climate disasters over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='TOTAL', data=climate_disasters_df)
plt.title('Total Number of Climate Disasters Over Time')
plt.xlabel('Year')
plt.ylabel('Total Disasters')
plt.grid(True)
plt.show()
```

Total Disaster Trend Analysis: The second line plot illustrates the total number of climate disasters recorded each year. This trend provides an overview of the overall increase or decrease in climate disasters over time.

```python
# Heatmap Analysis: Creating a heatmap for the dataset
heatmap_data = climate_disasters_df.set_index('Year')
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, cmap='coolwarm', annot=True, fmt=".0f")
plt.title('Heatmap of Climate Disasters Frequency by Year')
plt.ylabel('Year')
plt.xlabel('Disaster Type')
plt.show()
```

Heatmap Analysis: The heatmap visualizes the frequency of different types of climate disasters across years. This comprehensive view allows for easy comparison of the frequency and types of disasters across different years.e.

```python
# Distribution Analysis: Histograms for each disaster type
fig, axs = plt.subplots(2, 3, figsize=(15, 8))
for i, disaster in enumerate(disaster_types):
    row = i // 3
    col = i % 3
    sns.histplot(climate_disasters_df[disaster], kde=True, ax=axs[row, col])
    axs[row, col].set_title(f'Distribution of {disaster}')
plt.suptitle('Distribution of Different Types of Climate Disasters')
plt.tight_layout()
plt.show()
```

Distribution Analysis: The histograms show the distribution of each type of climate disaster. These plots are useful for understanding the frequency distribution and commonality of each disaster type.

```python
# Distribution Analysis: Histograms for carbon stocks and forest area
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
sns.histplot(forest_carbon_df['Carbon stocks in forests'], kde=True, ax=axs[0])
axs[0].set_title('Distribution of Carbon Stocks in Forests')
sns.histplot(forest_carbon_df['Forest area'], kde=True, ax=axs[1])
axs[1].set_title('Distribution of Forest Area')
plt.suptitle('Distribution Analysis of Carbon Stocks and Forest Area')
plt.show()
```

# Land Cover Accounts

Trend Analysis: The line plot illustrates the trends in different land cover categories over time, including "Total Land Cover Climate altering," "Total Land Cover Climate neutral," and "Total Land Cover Climate regulating." Index Analysis: The corrected line plot shows the trend in the "Climate Altering Land Cover Index Climate altering" over time. Distribution Analysis: The histograms depict the distribution of key variables, including the "Climate Altering Land Cover Index Climate altering" and total land cover in climate-altering and climate-neutral categories.
