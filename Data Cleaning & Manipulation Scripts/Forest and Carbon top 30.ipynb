```python
# Select specific columns
columns_to_keep = ['Country', 'Indicator'] + [f'F{year}' for year in range(1992, 2021)]
df_filtered = df[columns_to_keep]
```

```python
# Rename columns from 'FYear' to 'Year'
df_filtered.columns = ['Country', 'Indicator'] + [str(year) for year in range(1992, 2021)]
```

```python
# Melt the DataFrame to long format
df_long = pd.melt(df_filtered, id_vars=['Country', 'Indicator'], var_name='Year', value_name='Value')
```

```python
# Pivot the DataFrame to wide format based on 'Country', 'Year', and 'Indicator'
df_wide = df_long.pivot_table(index=['Country', 'Year'], columns='Indicator', values='Value', aggfunc='first').reset_index()
```

```python
df_filtered_countries = df_wide
```

```python
# Load the dataset
file_path = "C:/Users/ATAISH NEHRA/Downloads/Cleaned Datasets/Forest and Carbon top 30 clean.csv"
df = pd.read_csv(file_path)
```

```python
# Define the geographically closest countries to Afghanistan for imputation
closest_geographical_countries = ['Iran, Islamic Rep. of', 'Pakistan', 'China, P.R.: Mainland', 'India']
```

```python
# List of columns identified for imputation
columns_to_impute = ['Carbon stocks in forests', 'Index of carbon stocks in forests', 'Index of forest extent', 'Land area', 'Share of forest area']
```

```python
# Perform imputation for each column
for column in columns_to_impute:
    for year in df['Year'].unique():
        # Calculate the mean for the geographically closest countries for the given year and column
        mean_value_geographical_countries = df[(df['Country'].isin(closest_geographical_countries)) & (df['Year'] == year)][column].mean()
        
        # Only impute this mean value for Afghanistan for the same year where the value is NaN
        afghanistan_nan_rows = df[(df['Country'] == 'Afghanistan, Islamic Rep. of') & (df['Year'] == year) & (df[column].isna())]
        if not afghanistan_nan_rows.empty:
            df.loc[afghanistan_nan_rows.index, column] = mean_value_geographical_countries
```

```python
# Display the imputed dataset for Afghanistan
afghanistan_data_after_imputation = df[df['Country'] == 'Afghanistan, Islamic Rep. of']
afghanistan_data_after_imputation
```
