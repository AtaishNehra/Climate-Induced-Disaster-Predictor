```python
import pandas as pd
```

```python
# Select the required columns
columns_required = ['Country', 'Indicator', 'Climate_Influence', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020']
df_selected = df_original[columns_required]
```

```python
# Merge 'Indicator' and 'Climate_Influence' columns and remove the originals
df_selected['Indicator and Influence'] = df_selected['Indicator'] + ": " + df_selected['Climate_Influence']
df_selected.drop(['Indicator', 'Climate_Influence'], axis=1, inplace=True)
```

```python
df_selected
```

```python
# Melt the dataframe to long format for manipulation
df_long = df_selected.melt(id_vars=['Country', 'Indicator and Influence'], var_name='Year', value_name='Value')
```

```python
# Pivot the dataframe to get the desired structure
df_pivot = df_long.pivot_table(index=['Year', 'Country'], columns='Indicator and Influence', values='Value').reset_index()
```

```python
# Save the transformed dataset as a CSV file
output_file_path = 'C:/Users/ATAISH NEHRA/Downloads/Cleaned Datasets/Land Cover Accounts Top 30 clean.csv'
df_pivot.to_csv(output_file_path, index=False)
```
