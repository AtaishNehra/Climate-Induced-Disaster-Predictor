```python
import pandas as pd
```

```python
# Replace 'your_dataset.csv' with the path to your dataset
file_path = "C:/Users/ATAISH NEHRA/Downloads/DS5500 Raw datasets/Land cover accounts world only data.csv"
```

```python
# Load the dataset
data = pd.read_csv(file_path)
```

```python
# Print the column names
print(data.columns.tolist())
```

```python
# List of columns to keep
columns_to_keep = ['Indicator', 'Climate_Influence', 
                   '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', 
                   '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', 
                   '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', 
                   '2016', '2017', '2018', '2019', '2020']
```

```python
# Filter the DataFrame to keep only the specified columns
data_filtered = data[columns_to_keep]

# data_filtered now contains only the columns specified
```

```python
data_filtered
```

```python
# Melt the DataFrame to long format
long_format = data_filtered.melt(id_vars=['Indicator', 'Climate_Influence'], 
                                 var_name='Year', 
                                 value_name='Value')
```

```python
# Pivot to get the desired structure
pivoted_data = long_format.pivot_table(index='Year', 
                                       columns=['Indicator', 'Climate_Influence'], 
                                       values='Value')
```

```python
# Resetting the index to get 'Year' as a column
pivoted_data.reset_index(inplace=True)
```

```python
pivoted_data
```

```python
pivoted_data.columns = [' '.join(col).strip() for col in pivoted_data.columns.values]
```

```python
# Specify the file path and name for your new CSV file
output_file_path = "C:/Users/ATAISH NEHRA/Downloads/Cleaned Datasets/Land Cover Accounts_cleaned.csv"

# Save the DataFrame to a CSV file
pivoted_data.to_csv(output_file_path, index=False)

# 'index=False' ensures that the DataFrame's index (in this case, the 'Year') is not written as a separate column
```
