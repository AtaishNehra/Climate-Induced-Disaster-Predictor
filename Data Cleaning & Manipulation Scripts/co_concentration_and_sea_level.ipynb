```python
import pandas as pd
```

```python
sea_levels = pd.read_csv('./Raw datasets/Change_in_Mean_Sea_Levels.csv')
sea_levels.head()
```

```python
print(sea_levels.columns)
```

```python
#print(sea_levels.dtypes)
dates = sea_levels['Date'].str.split('/', expand = True)
sea_levels['Year'] = dates[2]
sea_levels.head()
```

```python
sea_levels = sea_levels.drop(columns = ['ObjectId', 'Country', 'ISO2', 'ISO3', 'Indicator', 'Unit', 'Source', 'CTS_Code', 'CTS_Name', 'CTS_Full_Descriptor', 'Date'])
#sea_levels = sea_levels['Value'].fillna(0)
sea = sea_levels.pivot_table(index = 'Year', columns = 'Measure', values = 'Value', aggfunc = 'mean').reset_index(drop=False)
sea.head()
```

```python
co_concentration = pd.read_csv('./Raw datasets/Atmospheric_CO%E2%82%82_Concentrations.csv')
co_concentration.head()
```

```python
print(co_concentration.columns)
```

```python
year_month = co_concentration['Date'].str.split('M', expand = True)
co_concentration['Year'] = year_month[0]
co_concentration['month'] = year_month[1]
co_concentration = co_concentration.drop(columns = ['ObjectId', 'Country', 'ISO2', 'ISO3', 'Indicator', 'Source', 'CTS_Code', 'CTS_Name', 'CTS_Full_Descriptor','Date'])
co_concentration.head()
```

```python
co_concentration['Year'] = co_concentration['Year'].astype(int)
co_concentration['month'] = co_concentration['month'].astype(int)
co_concentration = co_concentration[co_concentration.Year >= 1992]
co_concentration.head()
```

```python
co_concentration = co_concentration.drop(columns = ['month'])
co_concentration = co_concentration.pivot_table(index = 'Year', columns = 'Unit', values = 'Value', aggfunc = 'mean')
co_concentration.head()
```

```python
co_concentration = co_concentration.rename(columns = {'Parts Per Million': 'CO PPM', 'Percent':'CO Percent'}).reset_index(drop=False)
co_concentration.head()
```

```python
sea.to_csv('./Clean Datasets/sea_level_changes.csv', index = False)
co_concentration.to_csv('./Clean Datasets/co_concentrations.csv', index = False)
```
