```python
# Define the columns to keep
columns_to_keep = ['Indicator', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000', 
                   '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', 
                   '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020']
```

```python
# Filter the DataFrame to keep only the specified columns
df_filtered = df[columns_to_keep]
```

```python
# Display the first few rows of the filtered DataFrame
df_filtered.head(6)
```

```python
# Assuming df_filtered is your current DataFrame
# Pivot the DataFrame
df_pivoted = df_filtered.pivot_table(values=['1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000', 
                                             '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', 
                                             '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', 
                                             '2019', '2020'], 
                                     index='Indicator', 
                                     aggfunc='first').transpose()
```

```python
# Reset the index to make the years a column
df_pivoted.reset_index(inplace=True)
```

```python
df_pivoted.rename(columns={'index': 'Year'}, inplace=True)
```

```python
# Display the first few rows of the pivoted DataFrame
df_pivoted
```

```python
# Save the pivoted DataFrame to a CSV file
df_pivoted.to_csv("C:/Users/ATAISH NEHRA/Downloads/Cleaned Datasets/Forest and Carbon_cleaned.csv", index=False)
```
