```python
#creating the attention neural network
import tensorflow as tf
import keras
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#loading in data
sea_level = pd.read_csv('country_sea_levels.csv')
surface_temp = pd.read_csv('surface_temp_clean.csv')
forest = pd.read_csv('Forest and Carbon top 30 clean imputed.csv')
land_cover = pd.read_csv('Land cover top 30 cleaned.csv')
disasters = pd.read_csv('climate_disasters_clean.csv')
```

```python
#getting only values from 1992-2020
sea_level = sea_level[sea_level.Year <=2020]
disasters = disasters[disasters.Year >= 1992]
disasters = disasters[disasters.Year <=2020]
surface_temp = surface_temp[surface_temp.Year >=1992]
surface_temp = surface_temp[surface_temp.Year <= 2020]


#disasters = disasters.iloc[:, []]

#getting only the specific countries for disasters and surface temp
countries = ['World', 'United States', 'China, P.R.: Mainland', 'India', 'Philippines', 'Indonesia', 'Bangladesh', 'Vietnam', 'Mexico', 'Japan',
            'Brazil', 'Australia', 'France', 'Pakistan', 'Russian Federation', 'Thailand', 'Colombia', 'Afghanistan', 'Peru', 'Canada',
            'Haiti', 'Iran, Islamic Rep. of', 'Iran', 'Italy', 'Sri Lanka', 'South Africa', 'Argentina', 'Korea, Rep. of', 'Korea', 'Nepal', 'United Kingdom', 'Spain', 'Malaysia']
disasters = disasters.loc[disasters['Country'].isin(countries), :].reset_index()
surface_temp = surface_temp.loc[surface_temp['Country'].isin(countries), :].reset_index()

#changing column names to reflect their value meaning
sea_level.rename(columns = {'Value': 'Sea Level'}, inplace = True)
disasters.rename(columns = {'TOTAL': 'Total Disasters'}, inplace = True)
```

```python
#initializing hyper parameters
epoch = 100
batch = 100

#Getting X and Y for the different sets
'''X_train = train.drop(['Drought', 'Extreme temperature', 'Flood', 'Landslide', 'Storm', 'Total Disasters', 'Wildfire'], axis=1)
X_test = test.drop(['Drought', 'Extreme temperature', 'Flood', 'Landslide', 'Storm', 'Total Disasters', 'Wildfire'], axis=1)
X_vali = vali.drop(['Drought', 'Extreme temperature', 'Flood', 'Landslide', 'Storm', 'Total Disasters', 'Wildfire'], axis=1)

Y_train = train[['Year','Country','Total Disasters']]
Y_test = test[['Year','Country','Total Disasters']]
Y_vali = vali[['Year', 'Country', 'Total Disasters']]'''

X_train = train.drop(['Country','Drought', 'Extreme temperature', 'Flood', 'Landslide', 'Storm', 'Total Disasters', 'Wildfire'], axis=1)
X_test = test.drop(['Country','Drought', 'Extreme temperature', 'Flood', 'Landslide', 'Storm', 'Total Disasters', 'Wildfire'], axis=1)
X_vali = vali.drop(['Country','Drought', 'Extreme temperature', 'Flood', 'Landslide', 'Storm', 'Total Disasters', 'Wildfire'], axis=1)

Y_train = train[['Total Disasters']].astype(np.float32)
Y_test = test[['Total Disasters']].astype(np.float32)
Y_vali = vali[['Total Disasters']].astype(np.float32)
print(X_train.shape)
print(Y_train.shape)
```

```python
#using keras to create an initial attention nn
def attnnk(head, dim, unit, shape):
    model = tf.keras.models.Sequential()
    #model.add(tf.keras.layers.LSTM(unit, return_sequences = True, input_shape = (shape[1],shape[2])))
    model.add(tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(unit, return_sequences = True, input_shape = (shape[1],shape[2]))))
    model.add(tf.keras.layers.Dense(units = unit, activation = 'ReLU'))
    model.add(tf.keras.layers.Dense(units = unit, activation = 'ReLU'))
    model.add(tf.keras.layers.Dropout(rate = .2))
    query = tf.keras.layers.Input(shape = (shape[1], shape[2]))
    key = tf.keras.layers.Input(shape = (shape[1], shape[2]))
    value = tf.keras.layers.Input(shape = (shape[1], shape[2]))
    #model.add(tf.keras.layers.MultiHeadAttention(num_heads = head, key_dim = dim, value_dim = dim)([query, key,value]))
    attention = tf.keras.layers.MultiHeadAttention(num_heads = head, key_dim = dim, value_dim = dim)(query = query, key = key,value = value)
    model.add(tf.keras.layers.Dense(units = unit, activation = 'ReLU'))
    model.add(tf.keras.layers.Dense(units = 1, activation = 'ReLU'))
    model.compile(optimizer = 'Adam', loss = 'MeanSquaredError', metrics = ['Accuracy', 'AUC', 'RootMeanSquaredError'])
    return model
```

```python

X_train = X_train.values
X_train = X_train.reshape(X_train.shape[0], 1, X_train.shape[1])
X_test = X_test.values
X_test = X_test.reshape(X_test.shape[0], 1, X_test.shape[1])
model = attnnk(31, 64, 31, X_train.shape)
history = model.fit(X_train, Y_train, epochs=epoch, batch_size = batch)
results = model.evaluate(X_test, Y_test)
```
